***Importing KMeans for Scikit-Learn***

```python
from sklearn.cluster import KMeans
```

```python
MSS = []
for i in range(1,12):
  k_clusters = KMeans(n_clusters = i)
  k_clusters.fit(Iris_data.iloc[:,:-1])
  MSS.append(k_clusters.inertia_)
```

```python
plt.plot(range(1,12),MSS)
plt.xlabel('Number of Clusters')
plt.ylabel('Minimum Sum of Square')
plt.title('The Elbow Method')
```

***Making instance of Model Class***

```python
Kmeans = KMeans(n_clusters = 3)
```

***Fitting the data in the model and predicting the values also storing the predictions in a variable***

```python
y_predict = Kmeans.fit_predict(Iris_data.iloc[:,:-1])
```
