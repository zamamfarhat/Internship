# **Data Cleaning**

```python
gt.drop('eventid', axis=1, inplace  = True)
```

```python
gt.head(3)
```

```python
gt.shape
```

```python
pd.DataFrame(gt.isna().sum()).T
```

```python
thresold = len(gt)*0.5
gt = gt.dropna(thresh=thresold, axis=1)
```

```python
pd.DataFrame(gt.isnull().sum()).T
```

```python
gt.shape
```

```python
list_col  = gt.columns.to_list()
```

```python
for i in list_col:
  print(gt[i].describe())
  print("--------------")
```

```python
list_num_cols  = gt.select_dtypes(include = ['float64']).columns.to_list()
```

```python
for i in list_num_cols:
  gt[i].fillna(gt[i].mean(), inplace  =True)
```

```python
list_str_cols  = gt.select_dtypes(include = ['object']).columns.to_list()
```

```python
for i in list_str_cols:
  gt[i].fillna(gt[i].mode(), inplace  =True)
```
